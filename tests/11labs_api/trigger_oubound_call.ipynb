```python
import requests
import os
from dotenv import load_dotenv

load_dotenv("../../.env")

API_KEY = os.getenv("ELEVEN_LABS_API_KEY")
AGENT_ID = os.getenv("ELEVEN_LABS_AGENT_ID")
PHONE_NUMBER_ID = os.getenv("ELEVEN_LABS_AGENT_PHONE_NUMBER_ID")
TO_NUMBER = os.getenv("ALINA_PHONE_NUMBER")

url = "https://api.elevenlabs.io/v1/convai/twilio/outbound-call"

headers = {
    "Content-Type": "application/json",
    "xi-api-key": API_KEY,
}

payload = {
  "agent_id": AGENT_ID,
  "agent_phone_number_id": PHONE_NUMBER_ID,
  "to_number": TO_NUMBER,
  "conversation_initiation_client_data": {
    "conversation_config_override": {
      "agent": {
        "first_message": "Hi Alina — I’m calling about your outstanding balance.",
        "prompt": {"prompt": "Alina owes $10,000. Ask her politely to arrange payment."}
      }
    }
  }
}


response = requests.post(url, headers=headers, json=payload)
print(response.status_code)
print(response.json())
```

```python
import time

conversation_id = response.json()["conversation_id"]

while True:
    conv_response = requests.get(
        f"https://api.elevenlabs.io/v1/convai/conversations/{conversation_id}",
        headers={"xi-api-key": API_KEY},
    )
    status = conv_response.json()["status"]
    print(f"Status: {status}")

    if status in ("done", "failed"):
        break

    time.sleep(5) 

print(conv_response.json())
```
